# Regra IQR e Limites

```python
Q1 = df["Valor_Compra"].quantile(0.25)
Q3 = df["Valor_Compra"].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

print("Limite inferior:", limite_inferior)
print("Limite superior:", limite_superior)

df[(df["Valor_Compra"] < limite_inferior) | (df["Valor_Compra"] > limite_superior)]
```

# Coefieciente de variação

```python
cv_valor = df["Valor_Compra"].std() / df["Valor_Compra"].mean()
cv_idade = df["Idade"].std() / df["Idade"].mean()
cv_itens = df["Itens_Comprados"].std() / df["Itens_Comprados"].mean()

print("CV Valor_Compra:", cv_valor)
print("CV Idade:", cv_idade)
print("CV Itens_Comprados:", cv_itens)
```
